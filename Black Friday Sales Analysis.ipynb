```python

sales_by_category = df.groupby(
    "product_category"
)["purchase_amount"].sum()

sales_by_category.plot(kind="bar")

plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

```python
sales_by_city = df.groupby(
    "city"
)["purchase_amount"].sum()

sales_by_city.plot(kind="bar")

plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

```python
sales_by_payment = df.groupby(
    "payment_method"
)["purchase_amount"].sum()

sales_by_payment.plot(kind="bar")

plt.title("Sales by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Total Sales")

plt.show()
```

```python
sales_by_month = df.groupby(
    df["purchase_date"].dt.month
)["purchase_amount"].sum()

ax = sales_by_month.plot(kind="bar")

plt.title("Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")

ax.ticklabel_format(style="plain", axis="y")

plt.show()
```

```python
quantity_by_category = df.groupby(
    "product_category"
)["quantity"].sum()

quantity_by_category.plot(kind="bar")

plt.title("Quantity Sold by Category")
plt.xlabel("Product Category")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)

plt.show()
```
